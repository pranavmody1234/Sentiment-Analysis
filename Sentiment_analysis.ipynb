```python
def missing_values_analysis(df):
    na_columns_ = [col for col in df.columns if df[col].isnull().sum() > 0]
    n_miss = df[na_columns_].isnull().sum().sort_values(ascending=True)
    ratio_ = (df[na_columns_].isnull().sum() / df.shape[0]*100).sort_values(ascending=True)
    missing_df = pd.concat([n_miss, np.round(ratio_,2)], axis = 1, keys=['Missing Values', 'Ratio'])
    missing_df = pd.DataFrame(missing_df)
    return missing_df

def check_dataframe(df, head=5, tail=5):
    print("SHAPE".center(82,'~'))
    print('ROWS: {}'.format(df.shape[0]))
    print('Columns: {}'.format(df.shape[1]))
    print("TYPES".center(82, '~'))
    print(df.dtypes)
    print("".center(82,'~'))
    print(missing_values_analysis(df))
    print('DUPLICATED VALUES'.center(83,'~'))
    print(df.duplicated().sum())
    print("QUANTILES".center(82,'~'))
    print(df.quantile([0,0.05,0.50,0.95,0.99,1]).T)
    
check_dataframe(df)
```

```python
def check_class(dataframe):
    nunique_df = pd.DataFrame({'Variable': dataframe.columns,
                               'Classes': [dataframe[i].nunique() \
                                          for i in dataframe.columns]})
    nunique_df = nunique_df.sort_values('Classes', ascending=False)
    nunique_df = nunique_df.reset_index(drop=True)
    return nunique_df

check_class(df)
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import plot

def categorical_variable_summary(df, column_name):
    # Define color constraints and expand them if needed
    unique_categories = df[column_name].nunique()
    color_list = ['#B34D22', '#EBE00C', '#1FEB0C', '#0C92EB', '#EB0CD5']
    
    # Ensure we have enough colors for all unique values
    while len(color_list) < unique_categories:
        color_list.extend(color_list)  # Repeat the list until we have enough colors
    color_list = color_list[:unique_categories]  # Trim to match exact number of categories
    
    # Create subplots with appropriate titles
    fig = make_subplots(rows=1, cols=2, subplot_titles=('Countplot', 'Percentage'),
                        specs=[[{"type": "xy"}, {"type": "domain"}]])

    # Bar plot (Countplot)
    value_counts = df[column_name].value_counts()
    categories = [str(i) for i in value_counts.index]
    counts = value_counts.values.tolist()
    
    # If there are no values, print an error message and return
    if len(categories) == 0:
        print(f"No data found for column: {column_name}")
        return
    
    fig.add_trace(go.Bar(
        x=categories,
        y=counts,
        text=counts,
        textfont=dict(size=14),
        name=column_name,
        textposition='auto',
        showlegend=False,
        marker=dict(
            color=color_list[:len(categories)],  # Match the colors to the categories
            line=dict(color='#DBE6EC', width=1)
        )),
        row=1, col=1
    )

    # Pie chart (Percentage)
    fig.add_trace(go.Pie(
        labels=categories,
        values=counts,
        textfont=dict(size=18),
        showlegend=False,
        marker=dict(colors=color_list[:len(categories)])),  # Adjusted to match length
        row=1, col=2
    )

    # Update layout settings
    fig.update_layout(title={'text': column_name, 'y': 0.9, 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top'},
                      template='plotly_white')

    # Plot the figure
    plot(fig)  # Using plot() instead of iplot()

```

```python
categorical_variable_summary(df, 'overall')
```

```python
rt = lambda x: re.sub("[^a-zA-Z]", ' ', str(x))
df["reviewText"] = df["reviewText"].map(rt)
df["reviewText"] = df["reviewText"].str.lower()
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
df[['polarity', 'subjectivity']] = df['reviewText'].apply(lambda Text:pd.Series(TextBlob(Text).sentiment))

for index, row in df['reviewText'].iteritems():
    
    score = SentimentIntensityAnalyzer().polarity_scores(row)
    
    neg = score['neg']
    neu = score['neu']
    pos = score['pos']
    
    
    if neg>pos:
        df.loc[index, 'sentiment'] = "Negative"
    elif pos>neg:
        df.loc[index, 'sentiment'] = "Positive"
    else:
        df.loc[index, 'sentiment'] = "Neutral"
```
